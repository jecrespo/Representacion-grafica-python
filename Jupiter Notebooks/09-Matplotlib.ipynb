```python
from sklearn import datasets
import matplotlib.pyplot as plt
import numpy as np
```

```python
iris = datasets.load_iris()
```

```python
plt.figure()
x = np.arange(iris.data.shape[0])
colors = [['red','green','blue']]
#plt.scatter(x,iris.data[:,0],s=10*iris.target)
plt.scatter(x,iris.data[:,0],c=iris.target)
#plt.scatter(x,iris.data[:,0],edgecolor=colors[iris.target])
plt.title(iris.feature_names[0])
```
