```python
x = np.linspace(-10,10,50)
y = np.linspace(-10,10,50)
```

```python
plt.plot(x,y)
```

```python
x.shape=[1,50]
y.shape=[50,1]
```

```python
print(x)
print(y)
```

```python
fase = np.arctan2(x,y)
plt.figure()
plt.imshow(fase)
plt.colorbar()
```

```python
plt.colormap()
```

```python
plt.figure()
plt.imshow(fase,cmap='hsv')
plt.colorbar()
```

```python
plt.figure()
fase[fase<2] = 2
```

```python
plt.imshow(fase,cmap='hsv')
plt.colorbar()
```
