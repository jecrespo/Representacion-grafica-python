```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange (1,11)
print(x)
y = np.random.randn(10)
plt.xlabel("eje x")
plt.ylabel("eje y")
plt.plot(x,y,"rp-")
plt.plot(x,y--2,"b--")
plt.plot(x,y**3,"k-.")
plt.title("Dos lineas de tendencia")
plt.legend(['roja','azul','negra'])
plt.show()
```

### Leyenda en el plot

```python
plt.figure()
plt.xlabel("eje x")
plt.ylabel("eje y")
plt.plot(x,y,"rp-",x,y--2,"b--",x,y**3,"k-.")
plt.title("Dos lineas de tendencia")
plt.legend(['roja','azul','negra'])
plt.show()
```

### Envolvente

```python
x = np.linspace(0,10,50)
y = 1/(1+x)
sin_x = np.sin(2*np.pi*x)
plt.figure()
plt.plot(x,y*sin_x,label="principal")
plt.plot(x,y,label="envolvente+")
plt.plot(x,-y,label="envolvente-")
plt.title("Dos lineas de tendencia")
plt.legend(['roja','azul','negra'])
plt.show()
```
