```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,10,100)
y = np.random.rand(x.shape[0])
```

```python
plt.figure()
plt.subplot(1,2,1)
plt.plot(x,y)
plt.title("Ruido")
plt.subplot(1,2,2)
plt.plot(x,y**2)
plt.title("Ruido al cuadrado")
```

```python
plt.figure()
plt.subplot(2,2,1)
plt.plot(x,y)
plt.title("Ruido")
plt.subplot(2,2,4)
plt.plot(x,y**2)
title = plt.title("$Ruido ^2$",size = 24)
plt.suptitle("Dos Gráficas")
```
