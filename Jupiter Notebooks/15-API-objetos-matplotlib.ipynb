```python
phi_rad = np.linspace(0,2*np.pi)
fig, ax = plt.subplots()
plt.plot(phi_rad)
ax.set_ylabel('Angulo en Radianes')
ax_grados = ax.twinx()
angulo_grados = 180*np.random.rand(100)
ax_grados.plot(angulo_grados,'gx')
ax_grados.set_ylabel('Angulo en grados')
ax_grados.legend('grad')
ax.legend('rad')
```
