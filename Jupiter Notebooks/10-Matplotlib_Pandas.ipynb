```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
iris = pd.read_csv("http://www.openwith.net/wp-content/uploads/2017/10/iris.csv")
```

```python
iris['target']=pd.Categorical(iris.species)
plt.figure()
plt.scatter('sepal_length','petal_length',c = iris.target.cat.codes,data=iris)
```
