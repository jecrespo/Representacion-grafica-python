```python
import matplotlib.pyplot as plt
import pandas as pd
from ipywidgets import interact
```

```python
df = pd.read_csv(
        "https://raw.githubusercontent.com/selva86/datasets/" +
        "master/BostonHousing.csv")
@interact
def filtra_por_precio(precio_min=40):
    return df[df.medv > precio_min]
```

```python
plt.figure()
@interact
def plot_columna(columna='medv'):
    plt.clf()
    df[columna].plot()
```
