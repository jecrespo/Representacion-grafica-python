## gr√°fica del dataset iris

```python
from sklearn import datasets
import matplotlib.pyplot as plt
import numpy as np
```

```python
iris = datasets.load_iris()
```

```python
iris.keys()
```

```python
print(iris['DESCR'])
```

```python
plt.figure()
plt.plot(iris.data[:,0],'r',label=iris['feature_names'][0])
plt.plot(iris.data[:,1],'g',label=iris['feature_names'][1])
plt.plot(iris.data[:,2],'b',label=iris['feature_names'][2])
plt.plot(iris.data[:,3],'k',label=iris['feature_names'][3])
plt.legend()
```

```python
plt.figure()
plt.plot(iris.data, 'o', linewidth = 5, alpha = 0.7)
plt.title("demo iris")
plt.legend(iris.feature_names)
```

```python
print(iris.target)
print(iris.target_names)
```

```python
plt.figure()
plt.plot(iris.data[iris.target == 0], 'o', linewidth = 5, alpha = 0.7) #solo los de un tipo de datos
plt.title("demo iris")
plt.legend(iris.feature_names)
```
