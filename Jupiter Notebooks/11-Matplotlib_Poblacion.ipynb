```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
url = 'https://python-graph-gallery.com/wp-content/uploads/gapminderData.csv'
data = pd.read_csv(url)
```

# elegimos los datos de 1982

```python
data82 = data[data.year==1982]
```

```python
data82
```

```python
plt.figure()
data82['continent'] = pd.Categorical(data82['continent'])
plt.scatter(data82.lifeExp,
           data82.gdpPercap,
           s=data82['pop']/200000,
           c=data82.continent.cat.codes)
```

```python
data_country = data[data.country=='Spain']
plt.figure()
data_country['continent'] = pd.Categorical(data_country['year'])
plt.scatter(data_country.lifeExp,
           data_country.gdpPercap,
           s=data_country['pop']/20000,
           c=data_country.year)
plt.colorbar()
```
